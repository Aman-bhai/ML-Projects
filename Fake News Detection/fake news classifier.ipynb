```python
import nltk
nltk.data.path.append('"E:\\nltk_data\packages\corpora\stopwords.zip"')
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords as s
import re
import nltk
nltk.download('wordnet')

wn=WordNetLemmatizer()
stop_words=s.words('english')
```

```python
import string
```

```python
def clean_text(text):
  remov_pun=''.join([c.lower() for c in text if c not in string.punctuation])
  tokens=re.split('\W+',remov_pun)
  text=[ wn.lemmatize(word) for word in tokens if word not in stop_words]
  return text

```

```python
df['clean_text']=df['text'].apply(lambda x:clean_text(x.lower()))
```

```python
df.head()
```

```python
df.drop(['title','author','id'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(analyzer=clean_text)
```

```python
x=cv.fit(df['text'])
```

```python
import pickle
pickle.dump(cv,open('cvtransform.pkl','wb'))
```

```python
x=cv.transform(df['text'])
```
